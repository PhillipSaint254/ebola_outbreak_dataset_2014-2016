## Visialize the growth of Ebola in the different countries using subplots

```python
countries = sorted(df["Country"].unique())
countries
```

```python
# Determine subplot layout
n_countries = len(countries)
ncols = 2
nrows = (n_countries + ncols - 1) // ncols
print(f"Number of countries: {n_countries}, Subplot layout: {nrows} rows x {ncols} columns")
```

```python
fig, axes = plt.subplots(
    nrows=nrows,
    ncols=ncols,
    figsize=(15, 20),
    sharex=False
)

axes = axes.flatten()

for ax, country in zip(axes, countries):
    country_data = df[df["Country"] == country]

    ax.plot(
        country_data["Date"],
        country_data["No. of confirmed cases"],
        marker='o',
        label='Confirmed Cases'
    )

    ax.plot(
        country_data["Date"],
        country_data["No. of confirmed deaths"],
        marker='x',
        label='Confirmed Deaths'
    )

    ax.set_title(country)
    ax.set_xlabel("Date")
    ax.set_ylabel("Number of Cases/Deaths")
    ax.legend()
    ax.grid(True)

# Remove any unused subplots
for i in range(len(countries), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

 ## Guinea case study

```python
guinea_data = df[df["Country"] == "Guinea"]
guinea_data.head()
```

```python
g1 = guinea_data.plot.scatter(x="Date", y="No. of confirmed cases", color="blue")
guinea_data.plot.scatter(x="Date", y="No. of confirmed deaths", color="red", ax=g1)
plt.title("Guinean Confirmed Cases vs Confirmed Deaths")
plt.ylabel("Confirmed Cases/Death")
plt.xlabel("Date")
plt.show()
```

#### Fatality Rate

```python
# We can look at the Case Fatality Rate for all our datapoints to see if there is a trend
guinea_data["CFR"] = (
    guinea_data["No. of confirmed deaths"] / guinea_data["No. of confirmed cases"]
) * 100
guinea_data["CFR"].plot()
plt.tight_layout()
plt.show()
```

```python
np.mean(guinea_data["CFR"])
```

```python
guinea_data.head(20)
```

```python
guinea_data.tail(20)
```

We can see that, over time, the case fatility rate is on a rise until around the 1000th mark where the curve starts to flatten 

```python
# Let's checkout the fatility rate against the number of confirmed cases to see if there is a relationship
plt.plot(guinea_data["No. of confirmed cases"], guinea_data["CFR"])
plt.title("CFR against confirmed cases")
plt.ylabel("CFR")
plt.xlabel("No. of confirmed cases")
plt.grid()
plt.show()
```

- We can see an averagely high fatality when the cases are low, at 500
- Towards the 1000 mark, the fatality rate decreases and stays the same until about 1200 cases
- From there, henceforth, we can see that the fatality rate increases to a peak of about 62%

```python
# Let's try to establish a trendline between fatility rate, confirmed cases and death rate
# Let's checkout the fatility rate against the number of confirmed cases to see if there is a relationship
plt.plot(guinea_data["No. of confirmed cases"], guinea_data["CFR"], color='b')
plt.plot(guinea_data['No. of confirmed deaths'], guinea_data["CFR"], color='r')
plt.title("CFR against confirmed cases")
plt.ylabel("CFR")
plt.xlabel("No. of confirmed cases")
plt.grid()
plt.show()
```

```python
# Lets look for the correlation between confirmed cases and deathrates
correlation = guinea_data[["No. of confirmed cases", "No. of confirmed deaths"]].corr()
correlation
```

- As expected, there is almost a 1 to 1 correlation between no of confirmed cases and deaths
